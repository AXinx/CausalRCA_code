```python
import time, datetime
import requests 
import pickle as pkl 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm, trange
import json
import networkx as nx

import warnings
warnings.filterwarnings('ignore')

from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_recall_fscore_support
```

```python
def query_prom(query, startTime, endTime, PROMETHEUS = 'http://51.210.158.194:31090/', debug=False):
    response = requests.get(PROMETHEUS + '/api/v1/query_range',
      params={'query': query,
              'start': startTime,
              'end': endTime,
              'step': '5s'})
    if debug:
        print(response.json())
    results = response.json()['data']['result']
    return results
```

```python
from varname import nameof

names = ['front-end', 'user', 'catalogue', 'orders', 'carts', 'payment', 'shipping']

service_latency = pd.DataFrame()
for name in names:
    try:
        ctn_latency = 'sum(rate(request_duration_seconds_sum{name="'+name+'"}[1m])) / sum(rate(request_duration_seconds_count{name="'+name+'"}[1m]))'
        query_exp = ctn_latency
        query_res = query_prom(query_exp, startTime, endTime, debug=False)[0]
        values = list(map(float, np.array(query_res['values'])[:,1]))
        if len(set(values)) != 1:
            values = pd.DataFrame(values, columns=[str(name+'_latency')])
            service_latency.reset_index(drop=True, inplace=True)
            values.reset_index(drop=True, inplace=True)
            service_latency = pd.concat([service_latency, values],axis=1)
    except:
        print('No value of this metric')
time = list(map(int, np.array(query_res['values'])[:,0]))
time_ax = [datetime.datetime.fromtimestamp(i) for i in time]
service_latency.insert(loc=0, column='time',value=time_ax)
```

## Pre-process

```python
#normalize data
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
normalize = pd.DataFrame(scaler.fit_transform(metrics_value.drop(['time'], axis=1)))
normalize.columns = metrics_value.columns[1:]
normalized_values = pd.merge(metrics_value.time, normalize, left_index=True, right_index=True, how='left')
normalized_values
```

```python
data = normalized_values.drop(['time'], axis=1)
data_dropd = data.T.drop_duplicates().T
data_dropd
```
